```python

param_grid_spectral_car = {
    'n_clusters': [2,3,4,],
}

param_grid_spectral_customer = {
    'n_clusters': [2,3,4,5],
}

# clustering the car dataset
spectral_scores_car = spectral_clustering(prep.prep_car, param_grid_spectral_car)

# clustering the customer dataset
spectral_scores_customer = spectral_clustering(prep.prep_customer, param_grid_spectral_customer)
```

```python
# show the params in each metric

for metric_name in spectral_scores_car:
    for i, (score, model, params) in enumerate(spectral_scores_car[metric_name]) :
        print(f"{metric_name}: {score}, {params}")

for metric_name in spectral_scores_customer:
    for i, (score, model, params) in enumerate(spectral_scores_customer[metric_name]) :
        print(f"{metric_name}: {score}, {params}")
```

```python
def plot_label_on_pca(label, data, fig_suffix, output=False, render=False):
    # concat the label to the data
    data = np.concatenate((data, label.reshape(-1, 1)), axis=1)
    

    fig, ax = plt.subplots()
    ax.scatter(data[:, 0], data[:, 1], c=label)
    ax.set_title(f"PCA of {fig_suffix} Data")
    ax.set_xlabel("PCA 1")
    ax.set_ylabel("PCA 2")
    if render:
        plt.show()
    if output:
        plt.savefig(f"figs/pca_plot_{fig_suffix}.png")

# spectral_scores_car["silhouette_score"][0][1].row_labels_
plot_label_on_pca(spectral_scores_car["silhouette_score"][1][1][0], prep.prep_car, "car")
plot_label_on_pca(spectral_scores_customer["silhouette_score"][0][1][0], prep.prep_customer, "customer")

```
